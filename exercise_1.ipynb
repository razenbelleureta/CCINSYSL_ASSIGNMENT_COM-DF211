1. Find the function **f(x)**  

| Input  | CAT | DOG | APPLE | BANANA | ORANGE |  
| --- | --- | --- | --- | --- | --- |
| Output | cat | dog | apple | banana | orange


```python
#Ureta
def lowercase(inputs):
  outputs = []
  for x in inputs:
    outputs.append(x.lower())
  return outputs

print('INPUTS:  ',(['CAT','DOG','APPLE', 'BANANA', 'ORANGE']))
print('OUTPUTS: ',lowercase(['CAT','DOG','APPLE', 'BANANA', 'ORANGE']))
```

2. Find the function **f(x)**  

| Input  | CAT | DOG | APPLE | BANANA | ORANGE |  
| --- | --- | --- | --- | --- | --- |
| Output | CT | DG | PPL | BNN | RNG


```python
given_inputs = ['CAT','DOG','APPLE', 'BANANA', 'ORANGE']
print('The Given Inputs ARE: ',given_inputs)
words = input('Please Enter a word: ')

def consonant(words):
    vowels=['a','e','i','o','u']
    cons = [ x for x in words if x not in vowels]
    for x in cons:
        print(x)

print('The Consonants is/are: ')
consonant(words)
```

4. Find the function **f(x)**  

| Input  | 2 | 3 | 4 | 5 | 6 |  
| --- | --- | --- | --- | --- | --- |
| Output | 4 | 9 | 16 | 25 | 36


```python
def double_number(inputs):
    return [i*i for i in inputs]

inputs = [2,3,4,5,6]
print('The given numbers are: ', inputs)
print('The doubled number of the number in the lists are: ',double_number(inputs))
```

5. Find the function **f(x)**  

| Input  | 64 | 27 | 1000 | 729 | 512 |  
| --- | --- | --- | --- | --- | --- |
| Output | 4 | 3 | 10 | 9 | 8

```python
inputs = [64, 27, 1000, 729, 512]

def cube(inputs):
  for x in inputs:
    cube = x ** (1/3)
    print(round(cube))

print('These are the cube root of the given inputs: ')
cube(inputs)
```
